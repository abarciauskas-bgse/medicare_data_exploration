## Historical Medicare Enrollment

[source: cms.gov/Research-Statistics-Data-and-Systems](https://www.cms.gov/Research-Statistics-Data-and-Systems/Statistics-Trends-and-Reports/CMSProgramStatistics/Downloads/Medicare_Enrollment_Trend_by_State_1985_2012.zip)

```python
import pandas
import locale
from locale import atoi
locale.setlocale(locale.LC_NUMERIC, '')

filename = 'medicare_data/Medicare_Enrollment_Trend_by_State_1985_2012.xlsx - Table 1.csv'
medicare_enrollment = pandas.read_csv(filename)
medicare_enrollment.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

medicare_enrollment_indexed = medicare_enrollment.set_index('AREA OF RESIDENCE')
medicare_enrollment_as_int = medicare_enrollment_indexed.loc['UNITED STATES'].map(atoi)

plt.plot(medicare_enrollment_as_int)
plt.xlabel('Year')
plt.ylabel('U.S. Medicare Enrollment in 10\'s of millions')
plt.show()
```
